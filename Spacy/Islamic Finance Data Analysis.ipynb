```python
def pdf_to_csv(filename):
    from cStringIO import StringIO  
    from pdfminer.converter import LTChar, TextConverter
    from pdfminer.layout import LAParams
    from pdfminer.pdfparser import PDFDocument, PDFParser
    from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter

    class CsvConverter(TextConverter):
        def __init__(self, *args, **kwargs):
            TextConverter.__init__(self, *args, **kwargs)

        def end_page(self, i):
            from collections import defaultdict
            lines = defaultdict(lambda : {})
            for child in self.cur_item._objs:                #<-- changed
                if isinstance(child, LTChar):
                    (_,_,x,y) = child.bbox                   
                    line = lines[int(-y)]
                    line[x] = child._text.encode(self.codec) #<-- changed

            for y in sorted(lines.keys()):
                line = lines[y]
                self.outfp.write(";".join(line[x] for x in sorted(line.keys())))
                self.outfp.write("\n")

    # ... the following part of the code is a remix of the 
    # convert() function in the pdfminer/tools/pdf2text module
    rsrc = PDFResourceManager()
    outfp = StringIO()
    device = CsvConverter(rsrc, outfp, codec="utf-8", laparams=LAParams())
        # becuase my test documents are utf-8 (note: utf-8 is the default codec)

    doc = PDFDocument()
    fp = open(filename, 'rb')
    parser = PDFParser(fp)       
    parser.set_document(doc)     
    doc.set_parser(parser)       
    doc.initialize('')

    interpreter = PDFPageInterpreter(rsrc, device)

    for i, page in enumerate(doc.get_pages()):
        outfp.write("START PAGE %d\n" % i)
        if page is not None:
            interpreter.process_page(page)
        outfp.write("END PAGE %d\n" % i)

    device.close()
    fp.close()

    return outfp.getvalue()
```

```python
pdf_to_csv("emily.pdf")
```

```python
# Create a string of text
article = "This sentence is a test positive negative political test"
print(article)
# split the text

articlearray = article.split()
print(articlearray)

# create counts for Positive/Neg/Soc/rel/Pol
positivearray=['positive', 'good', 'great', 'awesome', 'affirmative']
negativearray=['negative', 'bad', 'sad', 'abandon', 'abate']
socialarray=['social', 'people', 'ally', 'attachment', 'family']
religiousarray=['religious', 'angel', 'believer', 'bible', 'Quran']
politicalarray=['political', 'government', 'politics', 'agency', 'order']


#Generate word Counts
poscount=0
negcount=0
soccount=0
relcount=0
polcount=0

# for every word in text

for i in articlearray:
    for p in positivearray:
        if i == p:
            poscount+= 1;
            break
    for n in negativearray:
        if i == n:
            negcount+= 1;
            break
    for s in socialarray:
        if i == n:
            soccount+= 1;
            break
    for r in religiousarray:
        if i == r:
            relcount+= 1;
            break    
    for g in politicalarray:
        if i == g:
            polcount+= 1;
            break
  



print("Positive Word Count:", poscount)
print("Negative Word Count:", negcount)
print("Social Word Count:", soccount)
print("Religious Word Count:", relcount)
print("Political Word Count:", polcount)

print("Political Word Count:", polcount)
```
